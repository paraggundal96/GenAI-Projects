### Data Ingestion Techniques

* WebBaseLoader
* PyPDFLoader
* TextLoader
* Check more on https://reference.langchain.com/python/langchain-community

```python
from langchain_community.document_loaders import TextLoader
```

```python
loader = TextLoader('AI.txt')
text = loader.load()
```

### Text Data Splitting

* RecursiveCharacterTextSplitter
* CharacterTextSplitter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size = 100, chunk_overlap = 20)
docs = splitter.split_documents(text)

print(docs[0])
print(docs[1])
```

### Vector Embedding

```python
from langchain_ollama import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model = 'llama3:8b')
)
```

```python
embeddings.embed_query("What is My Name")
```

```python
embeddings.embed_query("My Name")
```

### VectorStore Database

```python
from langchain_community.vectorstores import FAISS
```

```python
embeddings = OllamaEmbeddings(model = 'llama3:8b')
db = FAISS.from_documents(docs,embeddings)
```

```python
context = db.similarity_search("AI and industry bridge ")
```

```python
context
```

### Retriever

```python
retriever = db.as_retriever()
context = retriever.invoke("Significance of AI")
context
```

 FAISS also allows similarity search with score, it uses Manhattan distance for difference in vector spaces hence lower score is best

```python
context_w_score = db.similarity_search_with_score('Significance of AI')
context_w_score
```
