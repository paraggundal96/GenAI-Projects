### Pydantic

* It converts input data -> structured objects and raises error if data doesn't match the schema
* Commonly Used in API and AI pipelines

For instance, in a web service, incoming JSON data representing a user (e.g., name, age, email) can be mapped to a Pydantic model. The library automatically validates data types, enforces required fields, and raises errors if the input is malformed, thereby ensuring data integrity before further processing.

```python
from pydantic import BaseModel, Field
from langchain.chat_models import init_chat_model
```

```python
from re import M


model = init_chat_model(model = 'mistral:7b', model_provider='ollama')
```

```python
class Actor(BaseModel):

    name:str = Field(..., description="Name of the Actor")
    role:str

class Movie(BaseModel):

    name:str = Field(..., description= "Name of the movie")
    year:int = Field(..., description="The year the movie was released")
    cast:list[Actor] 
    rating:float = Field(..., description="IMDB ratings to movie")
    budget:float|None = Field(None, description="Budget for movie in Crores INR")
```

```python
model_w_struct = model.with_structured_output(Movie)
```

```python
model_w_struct.invoke("Avengers")
```

```python
from typing import TypedDict, Annotated

class Actor(TypedDict):
    name: Annotated[str, "Name of the Actor"]
    role: str

class Movie(TypedDict):
    name: Annotated[str, "Name of the movie"]
    year: Annotated[int, "The year the movie was released"]
    cast: list[Actor]
    rating: Annotated[float, "IMDB ratings to movie"]
    budget: Annotated[float | None, "Budget for movie in Million USD"]
```

```python

model_w_struct = model.with_structured_output(Movie)

```

```python
model_w_struct.invoke("Avengers")
```
